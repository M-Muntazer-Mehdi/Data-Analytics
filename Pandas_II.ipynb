```python
sub = pd.DataFrame({
    "key":["a", "a", "c", "c"], 
    "data": pd.Series(range(4), dtype="Int32")
})
sub
```

```python
sub1 = pd.DataFrame({
    "key":["a", "a", "c", "b", "b", "c", "a"], 
    "data1": pd.Series(range(7), dtype="Int32")
})
sub1
```

```python
temp = pd.merge(sub, sub1, how="outer", left_on="key", right_on="key") #try left_on, right_on 
temp
```

```python
sub2 = pd.DataFrame({
    "key":["a", "b"], 
    "data":pd.Series([3, 6], dtype="Int32")
})
sub2

pd.merge(sub1, sub2, how="inner")
```

```python
sub3 = pd.DataFrame({ "data":[3, 6]},
                    index=["a", "b"])
sub3
```

```python
pd.merge(sub1, sub3, left_on="key", right_index=True)
```
