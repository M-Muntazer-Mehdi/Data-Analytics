#### Importing the modules

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(4,3)})
```

#### Starting with Scatter Plot

Scatter plot is another pairwise (x,y) plot for represent the pair on x-axis and y-axis

```python
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.choice(["Head", "Tail"], 20)
sns.relplot(x=x, y=y, hue=z)
plt.title("Random scatter plot")
plt.legend(loc="lower right")
```

```python
sns.set_theme(style="darkgrid") #white, dark, whitegrid, darkgrid
sns.color_palette("husl", 8)
```

```python
myplte = sns.color_palette("muted")
myplte
```

```python
tips = sns.load_dataset("tips")
print (tips.columns)
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker", palette="husl")
```

```python
ax = sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
# ax.set_legend()
```

#### Line Plot 
Another Pairwise plot type 

```python
x = np.linspace(0,10,100)
y = np.sin(x)
sns.relplot(x = x, y=y, kind="line", hue=2, marker="o", palette="husl")
# sns.setlegend(title="")
```

```python
sns.lineplot(x = np.linspace(0,10,100), y=np.cos(x), color = "#FF5733", linestyle="dashdot")
```

```python
dowjones = sns.load_dataset("dowjones")
sns.lineplot(data=dowjones, x="Date", y="Price", linestyle="dashed")
```

```python
healthexp = sns.load_dataset("healthexp").sort_values("Year")
sns.lineplot(data=healthexp, x="Spending_USD", y="Life_Expectancy", 
            hue="Country", palette = "muted", color=["m","r","b","y"], 
            sort=False)
plt.legend(loc='best')
```

```python
sns.set_theme(style="ticks")
dots = sns.load_dataset("dots")
print ("data columns\n\n")
print (dots.columns)

print ("data head\n\n")
print (dots.head(5))

print ("data tail\n\n")
print (dots.tail(5))
```

```python
palette = sns.color_palette("bright")

# Plot the lines on two facets
sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="align", col="choice",
    kind="line", size_order=["dots", "sacc"], palette=palette,
    height=4, aspect=.75, facet_kws=dict(sharex=False),
)
```

```python
# Generate random data
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(1, 10, len(categories))
sns.barplot(x=categories, y=values,
            palette="pastel")
sns.despine(left=True, bottom=True)
```

```python
sns.set_theme(style="whitegrid")
penguins = sns.load_dataset("penguins")

g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="muted", alpha=.8, height=4
)
g.despine(left=True)
plt.ylabel("Body Mass(g)")
```

```python
marks_ = [51, 42, 33, 67, 98, 102, 72, 65, 109]
sns.boxplot(marks_, orient="v", width=.15, color="#FFAA45", saturation=0.75)
sns.despine(offset=10, trim=True)
```

```python
sns.set_theme(style="ticks", palette="pastel")
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["r", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
```
